### Are there any noticeable trends in diamond and cubic zirconia prices over time, and can you attribute these trends to specific factors?

```python
def uni_plot(columns):
    plt.figure(figsize=(15,10))
    plt.subplot(2,2,1)
    sns.histplot(data=df,x=columns,kde=True,bins=20)
    plt.title('Histogram of {}'.format(columns))
    plt.subplot(2,2,2)
    sns.boxplot(data=df,x=columns)
    plt.title('Boxplot of {}'.format(columns))
    plt.show()
```

```python
for i in df .select_dtypes(exclude='object'):
    uni_plot(i)
```

```python
for col in df.select_dtypes(include='object'):
    sns.countplot(x=df[col])
    plt.show()
```

### What are the most popular combinations of cut, color, and clarity grades for both diamonds and cubic zirconia?

```python
# Get the unique values of each feature
cut_values = df['cut'].unique()
color_values = df['color'].unique()
clarity_values = df['clarity'].unique()

# Create a dictionary to store the most popular combinations of each feature
most_popular_combinations = {}

# Iterate over each feature and count the number of times each combination appears in the dataset
for cut_value in cut_values:
    for color_value in color_values:
        for clarity_value in clarity_values:
            combination_count = df[(df['cut'] == cut_value) & (df['color'] == color_value) & (df['clarity'] == clarity_value)].shape[0]
            if combination_count > 0:
                most_popular_combinations[(cut_value, color_value, clarity_value)] = combination_count

# Get the most popular combination for each feature
most_popular_cut = max(most_popular_combinations.items(), key=lambda x: x[1])[0][0]
most_popular_color = max(most_popular_combinations.items(), key=lambda x: x[1])[0][1]
most_popular_clarity = max(most_popular_combinations.items(), key=lambda x: x[1])[0][2]

# Print the most popular combination
print(f'The most popular combination of cut, color, and clarity grades for cubic zirconia is: {most_popular_cut}, {most_popular_color}, {most_popular_clarity}')
```
